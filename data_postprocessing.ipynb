```python
from ipynb.fs.full.data_preprocessing import send_post_processing_params
from ipynb.fs.full.cnn_model_w12 import send_predicted_noise
from scipy.special import logsumexp
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
import math

strain_mean_train, strain_std_train, strain_mean_test, strain_std_test = send_post_processing_params()
noise_predict, tstrain_data_test = send_predicted_noise()
```

```python
print(noise_predict.shape, tstrain_data_test.shape)
print(noise_predict.min(), noise_predict.max(), noise_predict.mean())
print(tstrain_data_test.min(), tstrain_data_test.max(), tstrain_data_test.mean())
print(strain_mean_train, strain_std_train, strain_mean_test, strain_std_test)
```

```python
model_noise_predict = noise_predict
```

```python
#Un-normalizing noise_predict

# Before un-normalization 
plt.plot(model_noise_predict[0])
plt.title('noise_predict Before inversing z-score')
plt.savefig('noise_predict_Before_inversing_z-score.png')
plt.show()

#Un-normalizing Data
for i in range(noise_predict.shape[0]):
    noise_predict[i] = (noise_predict[i] * strain_std_train) + strain_mean_train
plt.plot(noise_predict[0])
plt.title('noise_predict After inversing z-score')
plt.savefig('noise_predict_After_inversing_z-score.png')
plt.show()

plt.plot(model_noise_predict[0], label = 'Normalized')
plt.plot(noise_predict[0], label = 'Un-normalized')
plt.title('noise_predict before and after un-normalization')
plt.savefig('noise_predict_before_and_after_un-normalization.png')
plt.legend()
plt.show()
```

```python
model_tstrain_data_test = tstrain_data_test
```

```python
#Un-normalizing tstrain_data_test

# Before un-normalization 
plt.plot(model_tstrain_data_test[0])
plt.title('tstrain_data_test Before inversing z-score')
plt.savefig('tstrain_data_test_Before_inversing_z-score.png')
plt.show()

#Un-normalizing Data
for i in range(tstrain_data_test.shape[0]):
    tstrain_data_test[i] = (tstrain_data_test[i] * strain_std_test) + strain_mean_test
plt.plot(tstrain_data_test[0])
plt.title('tstrain_data_test After inversing z-score')
plt.savefig('tstrain_data_test_After_inversing_z-score.png')
plt.show()

plt.plot(model_tstrain_data_test[0], label = 'Normalized')
plt.plot(tstrain_data_test[0], label = 'Un-normalized')
plt.title('tstrain_data_test After inversing z-score')
plt.savefig('tstrain_data_test_before_and_after_un-normalization.png')
plt.legend()
plt.show()
```

```python
#eight-order Butterworth bandpass filter
N = 8
Wn = [0, 0.3] #Band-pass [low end, high end]
filter_type = "bp"
fs = Wn[1]/2
b, a = signal.butter(N, Wn, filter_type, analog=True)
w, h = signal.freqs(b, a)
butter_sos = signal.butter(N, Wn, filter_type, fs , output = 'sos')
```

```python
for idx in range(len(noise_predict)):
        noise_predict[idx] = signal.sosfilt(butter_sos, noise_predict[idx])
```

```python
plt.plot(noise_predict[0])
plt.title('noise_predict after 8th order Butterworth Filter')
plt.savefig('noise_predict_after_Butterworth_Filter.png')
plt.show()
```

```python
#Flattening noise_predict and original strain
#to subtract and plot

noise_predict = noise_predict.reshape(noise_predict.shape[0]*noise_predict.shape[1])
tstrain_data_test = tstrain_data_test.reshape(tstrain_data_test.shape[0]*tstrain_data_test.shape[1])
```

```python
#anti-imaging filter not needed because all samples are sampled at the same rate

#subtracting from full bandwidth strain
clean_strain = tstrain_data_test - noise_predict
plt.plot(clean_strain)
plt.title('Clean Strain')
plt.savefig('Clean_Strain.png')
plt.show()
```

```python
#Calculating SNR

P_signal = logsumexp(np.mean(clean_strain ** 2))
P_noise = logsumexp((np.mean(noise_predict) ** 2))
print(P_signal, P_noise)
```

```python
SNR_dB = 10*math.log10(P_signal/P_noise)
print('SNR_dB =',round(SNR_dB,2))
```

```python
## Calculating Power Spectral Denisty (PSD) and Amplitude Spectral Density (ASD)

PSD_data_clean = plt.psd(clean_strain, Fs = 4096, scale_by_freq=False)
PSD_data_clean = np.asarray(PSD_data_clean)
plt.xscale("log")
plt.title('PSD of Clean Strain')
plt.savefig('PSD_of_Clean_Strain.png')
```

```python
PSD_data_orig = plt.psd(tstrain_data_test.reshape(clean_strain.shape), Fs = 4096, scale_by_freq=False,)
PSD_data_orig = np.asarray(PSD_data_orig)
plt.xscale("log")
plt.title('PSD of Original Strain')
plt.savefig('PSD_of_Original_Strain.png')
```

```python
#Calculate PSD Difference between Original and Clean Strain

plt.plot(PSD_data_clean[1],np.asarray(PSD_data_orig[0])- np.asarray(PSD_data_clean[0]))
plt.xscale('log')
plt.title('PSD Difference')
plt.savefig('PSD_Difference.png')
```

```python
ASD_data_clean = np.sqrt(PSD_data_clean[0])
plt.semilogx(PSD_data_clean[1],np.log(ASD_data_clean))
plt.grid()
plt.xlabel('Frequency')
plt.ylabel('Amplitude Spectral Density (dB)')
plt.title('ASD of Clean Strain')
plt.savefig('ASD_of_Clean_Strain.png')
```

```python
ASD_data_orig = np.sqrt(PSD_data_orig[0])
plt.semilogx(PSD_data_orig[1],np.log(ASD_data_orig))
plt.grid()
plt.xlabel('Frequency')
plt.ylabel('Amplitude Spectral Density (dB)')
plt.title('ASD of Original Strain')
plt.savefig('ASD_of_Original_Strain.png')
```

```python
#Calculate ASD Difference between Original and Clean Strain

plt.plot(PSD_data_clean[1],np.asarray(ASD_data_orig)- np.asarray(ASD_data_clean))
plt.xscale('log')
plt.title('ASD Difference')
plt.savefig('ASD_Difference.png')
```

```python
ASD_ratio = np.asarray(ASD_data_orig)/np.asarray(ASD_data_clean)
```

```python
#Calculating ASD ratio between Original/Clean

plt.plot(ASD_ratio)
plt.xscale('log')
plt.title('Original/Clean ASD Ratio')
plt.savefig('original_to_clean_asd_ratio.png')
```
